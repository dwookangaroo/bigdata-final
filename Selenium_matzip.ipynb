```python
# 맛집별 대표이미지를 하나씩 가져온다
driver = webdriver.Chrome('chrome_driver/chromedriver.exe') # 여기에 크롬드라이브 다운로드 받은 경로를 입력한다.
driver.get("https://www.google.co.kr/imghp?hl=ko&ogbl")
driver.implicitly_wait(3)

for i in tqdm(range(66,len(matzip_df.index))):
    elem = driver.find_element_by_name("q")
    elem.clear()
    elem.send_keys(matzip_df['셀레니움자료'][i])
    elem.send_keys(Keys.RETURN)
    images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
    result=[]
    
    try:
        result.append(images[0].get_attribute('src'))
        string_data = matzip_df['식당명'][i]
        urlretrieve(result[0], 'C:/crawling_image/' + str(string_data) + '_mainpicture.jpg')
        
    except:
        print('download directly the picture of ' + str(string_data))
    
```

```python
hotel_df = pd.read_csv('last_hotel.csv', encoding = 'cp949')
hotel_df
del hotel_df['Unnamed: 0']
```

```python
# 호텔별 대표이미지를 하나씩 가져온다
driver = webdriver.Chrome('chrome_driver/chromedriver.exe') # 여기에 크롬드라이브 다운로드 받은 경로를 입력한다.
driver.get("https://www.google.co.kr/imghp?hl=ko&ogbl")
driver.implicitly_wait(3)

for data in tqdm(hotel_df['주소'][106:]):
    elem = driver.find_element_by_name("q")
    elem.clear()
    elem.send_keys(data)
    elem.send_keys(Keys.RETURN)
    images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
    result=[]
    
    try:
        result.append(images[0].get_attribute('src'))
        string_data = data.replace("서울특별시","")
        urlretrieve(result[0], 'C:/crawling_image/' + str(string_data) + '_mainpicture.jpg')
        
    except:
        print('download directly the picture of ' + str(string_data))
```
